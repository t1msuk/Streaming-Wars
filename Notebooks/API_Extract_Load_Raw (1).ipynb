```python
import requests
import pandas as pd
from sqlalchemy import create_engine
```

```python
url = 'https://api.tvmaze.com/shows?page=1'
```

```python
response = requests.get(url)
shows_data = response.json()
```

```python
df_shows = pd.json_normalize(shows_data)
```

```python
df_shows = df_shows[['id', 'name', 'language', 'genres', 'premiered', 'rating.average', 'network.name']]
```

```python
df_shows.columns = ['Show_ID', 'Show_Name', 'Language', 'Genres', 'Premiered', 'Rating', 'Network']
```

```python
df = pd.DataFrame(df_shows)
df
```
